###  **Basic Of Probability**

- #### Tossing a coin 10,000 times and calculating the experimental probability of heads and tails.

```python
from fractions import Fraction
import random

coin_outcome = ['Head', 'Tail']

trial = 10000

heads_count = 0
tail_count = 0

for i in range(trial):
    toss = random.choice(coin_outcome)
    if toss == 'Head':
        heads_count += 1
    else:
        tail_count += 1

prob_head = Fraction(heads_count, trial)
prob_tail = Fraction(tail_count, trial)

print("Sample Space :", coin_outcome)
print("Total Toss :", trial)
print("Number of Head :", heads_count)
print("Number of Tail :", tail_count)
print("Probability of Head :", prob_head)
print("Probability of Tail :", prob_tail)
```

- #### Rolling two dice and computing the probability of getting a sum of 7.

```python
from fractions import Fraction
from itertools import product

dice_outcome = list(product(range(1,7), repeat = 2))

print("\nDice Outcomes :\n", dice_outcome)

total_outcome = len(dice_outcome)

fav_dice = [pair for pair in dice_outcome if sum(pair) == 7]

prob_sum_7 = Fraction(len(fav_dice), total_outcome)

print("\nTotal Outcome", total_outcome)
print("Favorable Outcome (sum = 7) :", fav_dice)
print("probability of getting a sum of 7 :", prob_sum_7)


```

- #### Write a function to estimate the probability of getting at least one "6" in 10 rolls of a fair die.

```python
from fractions import Fraction

p_no_six_one = Fraction(5,6) # probability of not getting a 6 in one roll

p_no_six_ten = p_no_six_one ** 10  # probability of not getting a 6 in 10 roll (5/6)^10

p_at_least_one_six = 1 - p_no_six_ten  # probability of not getting at least one 6 [1 - (5/6)^10]

print("probability of not getting a 6 in one roll :", p_no_six_one)

print("Probability of not getting a 6 in 10 rolls :", p_no_six_ten)

print("Probability of getting at least one 6 :", p_at_least_one_six)

```

***The probability of not getting a 6 in one roll of a fair die is 5/6. The probability of not getting a 6 in 10 independent rolls is (5/6)^10. Therefore, the probability of getting at least one 6 is 1 - (5/6)^10.***

### **Conditional Probability And Bayes' Theorem**

A bag contains 5 red, 7 green, and 8 blue balls. A ball is drawn is put back into the bag. If this process is repeated 1000 times.

- The probability of drawing a red ball given that the previous ball was blue.

```python
from fractions import Fraction
import random

bag = ['Red'] * 5 + ['Green'] * 7 + ['Blue'] * 8

repetitions = 1000

previous_blue = 0
red_after_blue = 0

previous = random.choice(bag)

for i in range(repetitions):
    current = random.choice(bag)

    if previous == 'Blue' :
        previous_blue += 1
        if current == 'Red' :
            red_after_blue += 1

    previous = current

prob_red_given_blue = Fraction(red_after_blue, previous_blue)

print("Total Repetitions :", repetitions)
print("Pervious ball was Blue :", previous_blue)
print("Red after Blue :", red_after_blue)
print("P(Red | Previous was Blue):", prob_red_given_blue)

```

### **Random Variables And Discrete Probability**

Generate a sample of size 1000 from a discrete random variable with the following distribution:
   - P(X=1) = 0.25
   - P(X=2) = 0.35
   - P(X=3) = 0.40

Compute the empirical mean, variance, and standard deviation of the sample.

```python
import numpy as np

values = [1, 2, 3]

probabilities = [0.25, 0.35, 0.40]

sample = np.random.choice(values, size = 1000, p = probabilities)

mean = np.mean(sample)
variance = np.var(sample)
std_deviation = np.std(sample)

print("Sample Size :", len(sample))
print("Empirical Mean :", mean)
print("Empirical Variance :", variance)
print("Empirical Standard Deviation :", std_deviation)
```

### **Continuous Random Variables**

Simulates 2000 random samples from an exponential distribution with a mean of 5. Visualize the distribution using.
   - A histogram
   - A probability density function (PDF) overlay 
    


```python
import numpy as np 
import matplotlib.pyplot as plt

mean = 5
data = np.random.exponential(scale = mean, size = 2000)

x = np.linspace(0, max(data), 1000)
pdf = (1/mean) * np.exp(-x/mean) # Exponential PDF Formula

plt.hist(data, bins = 40, density = True)

plt.plot(x, pdf)

plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Exponential Distribution (Mean = 5)")

plt.show()
```

### **Central Limit Theorem**

Simulate the Central Limit Theorem by following these steps
   - Generate 10,000 random numbers from a uniform distribution.
   - Draw 1000 samples of size n = 30.
   - Calculate and Visualize the distribution of sample means.

```python
import numpy as np 
import matplotlib.pyplot as plt

uniform_data = np.random.uniform(0, 1, 10000)

sample_mean = []

for i in range(1000):
    sample = np.random.choice(uniform_data, size = 30)
    sample_mean.append(np.mean(sample))


# Plot for Uniform Distribution

plt.hist(uniform_data, bins = 40, density = True)
plt.title("Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()

# Plot for Distribution of Sample Means

plt.hist(sample_mean, bins = 40, density = True)
plt.title("Distribution of Sample Means (CLT)")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.show()


```
