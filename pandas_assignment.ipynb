## First Operation

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
    'Salary': [45000, 54000, 50000, 62000, 47000]
}

df = pd.DataFrame(data)

print("First Five Row of DataFrame: \n",df.head(5))

print()

print("Summary Statistics of Age and Salaries columns: \n",df.describe())

print()

average_salary_hr = df.groupby('Department')['Salary'].mean().loc['HR']
print("The Average Salary Of HR is: ", average_salary_hr)
```

## Second Operation

```python
df_bonus = df.assign(Bonus = df['Salary'] * 0.10)
print("The DataFrame After Adding New Column Of Bonus(10%): \n", df_bonus)
```

## Third Operation

```python
emp_sal_age = df[(df['Age'] >= 25) & (df['Age'] <= 30)]
print(emp_sal_age)
```

## Fourth Operation

```python
df_group = df.groupby('Department')['Salary'].mean()
print("Average Salaries Of Each Department is: \n",df_group)
```

## Fifth Operation


```python
sorted_df = df.sort_values(by = 'Salary', ascending = False)
print("Sorted DataFrame According to Salary is: \n", sorted_df)
```
