## First Operation

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

plt.plot(x, y, marker = 'o', linestyle = '-', color = 'r', label = 'Y-Axis')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Simple Line Plot')
plt.legend()
plt.show()
```

## Second Operation

```python
import matplotlib.pyplot as plt

student = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

plt.bar(student, marks, color = ['red', 'blue', 'green', 'yellow'])
plt.xlabel('Student')
plt.ylabel('Marks')
plt.title('Student Records')
plt.show()
```

## Third Operation

```python
import matplotlib.pyplot as plt

regions = ['North America', 'Europe', 'Asia', 'Other']
revenue = [45, 25, 20, 10]

plt.pie(revenue, explode = [0.1, 0.0, 0.0, 0.0], labels = regions, autopct = '%1.1f%%')
plt.axis('equal')
plt.title('Revenue Distribution')
plt.show()
```

## Forth Operation


```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randint(1, 101, 1000)

plt.hist(data, bins = 20, edgecolor = 'black', color = 'red')

plt.title('Histogram Of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
