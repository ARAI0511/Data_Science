## First Operation

```python
import numpy as np 

arr = np.arange(1,21)
print("Original Array is: ",arr)
print("Additional of Array is: ",np.sum(arr))
print("Mean of Array is:",np.mean(arr))
print("Meadian of Array is: ", np.median(arr))
print("Standard Derivation of Array is: ", np.std(arr))

print()

print("The Element Greater than 10 in array is: ", arr[arr > 10])
```

## Second Operation

```python
arr_2d = np.arange(1,17).reshape(4,4)
print("2D Array is: ",arr_2d)

print()

print("Transpose of 2D Array is: ", arr_2d.T)

print()

print("Row Wise Calculation of 2D Array is: ", arr_2d.sum(axis = 1))
print("Column Wise Calculation of 2D Array is: ", arr_2d.sum(axis = 0))
```

## Third Operation

```python
arr_a = np.random.randint(1, 21, (3,3))
arr_b = np.random.randint(1, 21, (3,3))
print("Array A is: ", arr_a)
print("Array B is: ", arr_b)

print()

print("Adittion of Array A and Array B is: ", arr_a + arr_b)
print("Subtraction of Array A and Array B is: ", arr_a - arr_b)
print("Multiplication of Array A and Array B is: ", arr_a * arr_b)

print()

print("Dot Product of Array A and Array B is: ", arr_a.dot(arr_b))
```

## Fourth Operation

```python
arr_1d = np.arange(1,13)
print("1D Array is :", arr_1d)

print()

new_arr = arr_1d.reshape(3,4)
print("Reshaped Array is: ", new_arr)

print()

slice = new_arr[:2,-2:]
print("Array after slicing first two rows and last two columns is: ", slice)
```
